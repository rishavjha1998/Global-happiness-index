# Model development

```python
X=df[['GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices']]
```

```python
Y=df[['Score']]
Y.head(4)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
X_train
```

```python
lm.fit(X_train,y_train)
```

```python
y_hat=lm.predict(X_test)
```

```python
lm.coef_
```

```python
lm.intercept_
```

# model evaluation#

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE=mean_squared_error(y_test,y_hat)
print(MSE)
```

```python
from sklearn.metrics import r2_score
```

```python
R_square=r2_score(y_test,y_hat)
```

```python
print(R_square)
```

```python
import statsmodels.api as sm
```

```python
X_train = sm.add_constant(X_train)
```

```python
model = sm.OLS(y_train,X_train).fit()
```

```python
model.params
```

 Y=  1.842849 + 0.848085 * GDP per capita  + 1.279114 *Social support   + 0.881396*Healthy life expectancy + 1.640495*Freedom to make life choices 

```python
print(model.rsquared_adj)
```

```python
# This indicates that this is a good model
```
